```python
#1. Reverse a list
def list_reverse(user_input):
    
    if(type(user_input) is not list):
        raise TypeError("Enter list a list as your argument")
    
    if(len(user_input) == 0):
        raise ValueError("Please ensure to enter at least a number in the list")
    else:
        the_list = []      
        for i in user_input[::-1]:
            the_list.append(i)
    
    return the_list

#2. Check if the list is in increasing order
#3 different types of sequences: Strings (chars), Lists, Tuples
def is_sorted(seq):
    #Check for Tuple, Check for dictionary?
    if(type(seq) is str):
        if(not seq):
            raise ValueError("The sequence is empty. Ensure to have at least one character in the string")
            
        for x in range(len(seq)-1):
            if (seq[x] < seq[x+1]):
                continue
            else:
                return False
        return True
    
    if(type(seq) is list):
        if(not seq):
            raise ValueError("The sequence is empty. Ensure to have at least one element in the list")
        for i in range(len(seq)):
            if(i+1 > len(seq)-1):
                break
            if(seq[i]>seq[i+1]):
                return False
        return True
    if(type(seq) is tuple):
        if(len(seq) == 0):
            raise ValueError("The sequence is empty. Ensure to have at least one element in the tuple")
        
        try:
            for i in range(len(seq)):
                if(i+1 > len(seq)-1):
                    break
                if(seq[i]>seq[i+1]):
                    return False
            return True 
        except:
            raise TypeError("The sequence is not recognized. If this is a tuple of tuples ensure to have just a tuple")
            
    else:
        raise TypeError("The sequence is not recognized. Ensure to have a string, list, or tuple as your sequence")

#3. Binary Search given a sorted list 
def binary_search(t, elmt):
    
    #Check for basic errors
    #n is not an int
    if(type(elmt)is not int):
        raise TypeError("Error ensure elmt is of type int")
        
    #list is empty
    if(not t):
        raise ValueError("Error ensure that the list is not empty")
    #Only one element
    if(len(t) ==1):
        if(t[0] ==elmt):
            return True
        else:
            return False
    
    #Algorithm 
    firts_half = 0
    second_half = len(t)-1
    
    #Loop until the two "poles" do not pass eachother
    while (second_half>=firts_half):
        #mid point will be the min and max values divide by 2(floor)
        mid_point = (firts_half+second_half)//2  
        
        #If we find it 
        if(t[mid_point] == elmt):
            return True
        
        #Keep checking to the left 
        if(t[mid_point] > elmt):
            second_half= mid_point-1
            
        #Keep checking to right 
        if(t[mid_point]<elmt):
            firts_half= mid_point+1
            
    #At this point the element was not found
    return False
```

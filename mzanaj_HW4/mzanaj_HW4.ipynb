```python
#4 Fourth Problem, ~ <30 minutes
#1. HIGH ORDER FUNCTION 
#outer , give list 
def make_poly(coeffs):
    #inner, provide a number 
    def p(x):
        the_sum = 0
        #loop through elemen of the list, and find summation of each element times input to power (1 to n)
        for y in range(len(coeffs)):
            the_sum = the_sum + coeffs[y]*(x**y)
        return the_sum
    return p

#2. Polynomial evaluation
def eval_poly(coeffs, args):
    return list(map(make_poly(coeffs),args))
    
```

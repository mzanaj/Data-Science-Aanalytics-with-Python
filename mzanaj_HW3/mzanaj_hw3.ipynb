```python
#1. Biagrams from a file
import string
def count_bigrams_in_file(file_name):
    try:
        #Read File Contents
        s = ""
        f=open(file_name, 'r')

        #Read all words
        for line in f:
            for w in line.split():
                s= s+" "+ w

        #Strip all punctuations
        #Make lower case
        #Append all words into a list
        punct= string.punctuation
        l = []
        word = str()
        for x in s:
            if( x in punct):
                continue
            elif(x !=' '):
              
                word = word + x
            else:
                word = word.lower()
                l.append(word)
                word = str()
        #this was never executing given it is the last item
        l.append(word)

        l3 = []
        #Take out all spaces
        for x in range(len(l)):
            if '' == l[x]:
                continue
            else:
                l3.append(l[x])
       
        #Create Bigrams
        l2 = []
        for x in range(len(l3)-1):
            t1 = (l3[x],)
            t2= (l3[x+1],)
            t3 = t1+t2
            l2.append(t3)

        #Create key, value components
        the_dict = dict()
        for x in l2:
            #if not in there create one
            if x not in the_dict:
                the_dict[x] = 1
            #if already in there increase the current counter
            else:
                the_dict[x] +=1
                
        return the_dict
                
    except:
        #String type expected
        if(type(file_name) is not str):
            raise TypeError("Please ensure that the input is of type string")
        else:
            raise FileNotFoundError("Please provide a valid path!")

#2. Run function on WandP.Txt
import pickle
#Open file mb.bigrams.pickle.txt
file_biagrams= 'mb.bigrams.pickle'
file = open(file_biagrams, 'wb')
#Open WandP, (Make sure path is set to correct location)
file_wandp = 'WandP.txt'
#Close
file.write(pickle.dumps(count_bigrams_in_file(file_wandp)))
file.close()
```
